# Prewise Qwen3.5-4B — Vietnamese Context Adapters v3

Notebook huấn luyện hai LoRA riêng:

- `message-context-adapter`
- `explanation-adapter`

Chọn accelerator **GPU T4 x2**.

```python
from pathlib import Path
import json, subprocess, sys

candidates = list(Path('/kaggle/input').rglob('dataset_manifest.json'))
assert candidates, 'Không tìm thấy dataset_manifest.json trong /kaggle/input'
BUNDLE = candidates[0].parent
DATA = Path('/kaggle/working/prewise-data-v3')
WORK = Path('/kaggle/working/prewise-v3')
BASE_MODEL = 'Qwen/Qwen3.5-4B'

manifest = json.loads((BUNDLE/'dataset_manifest.json').read_text(encoding='utf-8'))
print('BUNDLE =', BUNDLE)
print(json.dumps(manifest['coverage'], ensure_ascii=False, indent=2))
```

## 1. Validate toàn bộ bundle

```python
subprocess.run([
    sys.executable,
    str(BUNDLE/'validate_vi_context_bundle_v3.py'),
    '--bundle-root', str(BUNDLE),
], check=True)
```

## 2. Giải nén dữ liệu

```python
subprocess.run([
    sys.executable,
    str(BUNDLE/'prepare_kaggle_data.py'),
    '--bundle-root', str(BUNDLE),
    '--output-root', str(DATA),
], check=True)
```

## 3. Cài dependency

```python
subprocess.run([
    sys.executable,
    str(BUNDLE/'install_kaggle_deps.py'),
], check=True)
```

## 4. Train hai adapter và đóng gói

```python
subprocess.run([
    sys.executable,
    str(BUNDLE/'run_kaggle_training.py'),
    '--bundle-root', str(BUNDLE),
    '--data-root', str(DATA),
    '--output-root', str(WORK),
    '--model-name', BASE_MODEL,
    '--message-train-samples', '32799',
    '--explanation-train-samples', '24000',
    '--validation-samples', '2400',
], check=True)
```

## 5. Kiểm tra file đầu ra

```python
archive = Path('/kaggle/working/prewise-adapters-v3.zip')
assert archive.exists(), archive
print('READY:', archive, archive.stat().st_size, 'bytes')
```
